```python
import pymongo
from pymongo import MongoClient
client = MongoClient()
client = MongoClient('localhost', 27017)
# client = MongoClient('mongodb://localhost:27017/')
db = client.db_cdiscount
```

```python
import numpy as np
import pandas as pd
```

```python
lst_encode_argmax = []
cur = db.cat_encode.find({})

while (cur.alive):
    idx = cur.next()
    dic = {}
    dic['cat'] = idx['cat']
    dic['arg_max'] = float(np.argmax(idx['encode']))
    
    lst_encode_argmax.append(dic)
lst_encode_argmax
```
